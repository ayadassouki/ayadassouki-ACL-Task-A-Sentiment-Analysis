````python
import re

def clean_text(text):
  # Remove special characters and punctuation
  text = re.sub(r"[^\w\s]", " ", text)

  # Remove single characters
  text = re.sub(r"\b[a-zA-Z]\b", " ", text)

  # Remove HTML tags
  text = re.sub(r"<[^>]*>", " ", text)

  # Lowercase the text
  text = text.lower()

  # Remove extra whitespace
  text = re.sub(r"\s+", " ", text)

  # Trim leading and trailing spaces
  text = text.strip()

  return text
````

````python
def clean_df(data):
  texts = data['text'].to_list()
  clean_texts = [clean_text(t) for t in texts]
  data['clean_text'] = clean_texts
  data.head()
````

````python
import re
import ast

# Step 1: Clean and preprocess the data
def clean_input(data):
    cleaned_data = []
    for item in data:
        # Remove unwanted characters like *, ```, \n, and \
        item = re.sub(r"[`\*\n\\]", "", item)
        item = item.strip()  # Remove leading/trailing spaces
        cleaned_data.append(item)
    return cleaned_data

# Step 2: Insert commas between key-value pairs and fix any invalid formatting
def fix_json_format(json_str):
    # Add missing commas between key-value pairs
    json_str = re.sub(r"(\"[^\"\}]+\"):\s*([\{\[\"-?\d]|true|false|null)", r"\1: \2, ", json_str)
    # Remove trailing commas before closing curly braces
    json_str = re.sub(r",\s*}", "}", json_str)
    return json_str



# Step 4: Clean the input data
cleaned_data = clean_input(result)

# Step 5: Join the cleaned list into a single string
joined_data = " ".join(cleaned_data)

# Step 6: Extract JSON-like blocks using regex
json_like_strings = re.findall(r"\{.*?\}", joined_data, re.DOTALL)


# Step 8: Output the parsed data
print(json_like_strings)
````

````python
def extract_emotions(data):
    extracted_data = []

    # Define the regex pattern to match key-value pairs for clean_text and emotion labels
    pattern = r"'clean_text':\s*'([^']*)'|('p_anger'|'p_fear'|'p_joy'|'p_sadness'|'p_surprise')\s*:\s*(\d)"

    for entry in data:
        # Initialize a dictionary to store extracted values
        extracted_dict = {}

        # Use regex to find matches
        matches = re.findall(pattern, entry)

        # Process matches to populate the dictionary
        for match in matches:
            if match[0]:  # This is for clean_text
                extracted_dict['clean_text'] = match[0]
            elif match[1] and match[2]:  # This is for emotion labels
                extracted_dict[match[1].strip("'")] = int(match[2])

        # Append the extracted data to the result list
        extracted_data.append(extracted_dict)

    return extracted_data

extracted_data = extract_emotions(json_like_strings)
extracted_data
````

````python
columns = ['p_anger', 'p_fear', 'p_joy', 'p_sadness', 'p_surprise']
for col in columns:
  # Replace non-finite values with 0 before converting to int
  merged_df[col] = merged_df[col].fillna(0).astype(int)

merged_df.head()
````

````python
from sklearn.metrics import classification_report

# Assuming you have true labels in the DataFrame as 'true_anger', 'true_fear', etc.
true_labels = ['Anger', 'Fear', 'Joy', 'Sadness', 'Surprise']
pred_emotion = ['p_anger', 'p_fear', 'p_joy', 'p_sadness', 'p_surprise']

# Create an empty dictionary to store classification reports for each emotion
classification_reports = {}

# Iterate through each emotion and calculate the classification report
for true_label, pred_label in zip(true_labels, pred_emotion):
    true_values = merged_df[true_label]  # True labels for the emotion
    pred_values = merged_df[pred_label]  # Predicted labels for the emotion


    # Generate the classification report
    report = classification_report(true_values, pred_values)

    # Store the report in the dictionary
    classification_reports[true_label] = report

# Print the classification reports
for emotion, report in classification_reports.items():
    print(f"Classification Report for {emotion}:")
    print(report)
    print("\n")
````
